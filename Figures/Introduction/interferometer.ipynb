```python
import numpy as np

import matplotlib.pyplot as plt
import rcparams
plt.rcParams.update(rcparams.params)

from matplotlib.patches import Rectangle, Wedge
```

```python
fig, ax = plt.subplots(dpi=180)

ax.set_aspect('equal')

ax.spines['left'].set_position('zero')
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_position('zero')
ax.spines['bottom'].set_visible(False)

ax.set_xticks([])
ax.set_yticks([])

ax.set_xlim(-4, 11)
ax.set_ylim(-4, 11)

# Lasers
# ------

laser_c = '#ff0000'
laser_lw = 3

ax.plot([-2, 0], [0, 0], c=laser_c, lw=laser_lw, zorder=10)
ax.plot(-1, 0, '>', c=laser_c, ms=6)

ax.plot([0, 0], [-2, 0], c=laser_c, lw=laser_lw, zorder=10)
ax.plot(0, -1, 'v', c=laser_c, ms=6)

split_laser_c = '#ff7373'
split_laser_lw = 1.5
split_laser_offset = 0.1

ax.plot([0, 10], [0+split_laser_offset, 0+split_laser_offset], c=split_laser_c, lw=split_laser_lw, zorder=10)
ax.plot(4, split_laser_offset, '>', c=split_laser_c, ms=3)

ax.plot([0, 10], [0-split_laser_offset, 0-split_laser_offset], c=split_laser_c, lw=split_laser_lw, zorder=10)
ax.plot(6, -split_laser_offset, '<', c=split_laser_c, ms=3)

ax.plot([0-split_laser_offset, 0-split_laser_offset], [0, 10], c=split_laser_c, lw=split_laser_lw, zorder=10)
ax.plot(-split_laser_offset, 4, '^', c=split_laser_c, ms=3)

ax.plot([0+split_laser_offset, 0+split_laser_offset], [0, 10], c=split_laser_c, lw=split_laser_lw, zorder=10)
ax.plot(split_laser_offset, 6, 'v', c=split_laser_c, ms=3)

# Mirrors
# -------

mirror_width = 0.2
mirror_height = 1.5

mirror_fc = '#d1d1d1'
mirror_ec = '#878787'

ax.add_artist(
    Rectangle(
        (10-mirror_width/2, 0-mirror_height/2), 
        mirror_width, 
        mirror_height,
        fc=mirror_fc,
        ec=mirror_ec,
        zorder=11
        )
    )

ax.add_artist(
    Rectangle(
        (0-mirror_height/2, 10-mirror_width/2), 
        mirror_height,
        mirror_width,
        fc=mirror_fc,
        ec=mirror_ec,
        zorder=11
        )
    )

# Beam splitter
# -------------

bs_height = 1.5
bs_width = 0.2

ax.add_artist(
    Rectangle(
        (0-bs_height/2, 0-bs_width/2), 
        bs_height,
        bs_width,
        angle=45,
        rotation_point='center',
        color='#545454',
        # fc=mirror_fc,
        # ec=mirror_ec,
        zorder=11
        )
    )

# Laser source
# ------------

laser_height = 0.5
laser_width = 1.5

ax.add_artist(
    Rectangle(
        (-2.8-laser_width/2, 0-laser_height/2), 
        laser_width,
        laser_height,
        fc='w',
        ec='k',
        zorder=11
        )
    )

# Photodetector
# -------------

opening_angle = 80

ax.add_artist(
    Wedge(
        (0,-2.9),
        1,
        90-opening_angle/2,
        90+opening_angle/2,
        fc='w',
        ec='k',
        zorder=11
        )
    )

# Text labels
# -----------

ax.text(10, -1.5, '$(L_x,0)$', horizontalalignment='center')
ax.text(-2.6, 10, '$(0,L_y)$', verticalalignment='center')
ax.text(-0.7, -0.7, '$(0,0)$', verticalalignment='top', horizontalalignment='right')

ax.text(10, 1.13, 'Mirror', horizontalalignment='center')
ax.text(1.1, 10, 'Mirror', verticalalignment='center')
ax.text(0.7, 0.7, 'Beam-splitter')
ax.text(-2.8, 0.6, 'Laser', horizontalalignment='center')
ax.text(0,-3.6, 'Photodetector', horizontalalignment='center')


fig.savefig('interferometer.pdf', bbox_inches='tight')
```
